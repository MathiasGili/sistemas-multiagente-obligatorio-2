```python
import numpy as np
from tqdm import tqdm

from games.kuhn3 import KuhnPoker3
from agents.counterfactualregret_t import CounterFactualRegret
from agents.ismcts import ISMCTS
from agents.minimax import MiniMax
from agents.agent_random import RandomAgent
```

```python
g = KuhnPoker3(play_order=[0, 1, 2])
```

```python
my_agents = {
    agent: CounterFactualRegret(game=g, agent=agent)
    for agent in g.agents
}

# Alternatives for quick experiments:
# my_agents[g.agents[0]] = RandomAgent(game=g, agent=g.agents[0])
# my_agents[g.agents[1]] = ISMCTS(game=g, agent=g.agents[1], simulations=200, rollouts=10, rollout_iterations=100)
```

```python
g.reset()
while not g.done():
    g.render()
    print(f"Agent {g.agent_selection}")
    action = my_agents[g.agent_selection].action()
    print(f"Action {action} - move {g.action_move(action)}")
    g.step(action)

g.render()
for agent in g.agents:
    print(f"Reward {agent} = {g.reward(agent)}")
```

```python
train_iterations = 1000

for agent in g.agents:
    print('Training agent ' + agent)
    if hasattr(my_agents[agent], 'train'):
        my_agents[agent].train(train_iterations)
        policies = {
            node: np.round(my_agents[agent].node_dict[node].policy(), 3)
            for node in sorted(my_agents[agent].node_dict.keys())
        }
        print(policies)
```

```python
cum_rewards = dict(map(lambda agent: (agent, 0.0), g.agents))
niter = 10000

for _ in tqdm(range(niter)):
    g.reset()
    while not g.done():
        action = my_agents[g.agent_selection].action()
        g.step(action=action)

    assert sum(g.rewards.values()) == 0, g.rewards
    for agent in g.agents:
        cum_rewards[agent] += g.rewards[agent]

avg_rewards = {
    agent: cum_rewards[agent] / niter
    for agent in g.agents
}
print('Average rewards:', avg_rewards)
print('Total average reward:', round(sum(avg_rewards.values()), 6))
```
